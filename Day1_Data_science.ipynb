# **1. Import Pandas & Create DataFrame**

```python
import pandas as pd

data1 = {
     'Name': ['Alice', 'Bob', 'Charlie', 'Pushpank', 'wiki', 'vivek'],
     'Age': [25, 30, 35, 23, 24, 25],
     'City': ['New York', 'San Francisco', 'Los Angeles', 'Bangalore', 'Jharkhand', 'Bokaro'],
     'department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'IT'],
     'salary': [20000, 30000, 40000, 50000, 60000, 70000]
}
```

```python
df = pd.DataFrame(data1)
df
```
